```python
%matplotlib inline

import pandas as pd

GITHUB = 'https://raw.githubusercontent.com'
FEATURES = '/phoible/phoible/master/data/FEATURES/phoible-segments-features.tsv'

features = pd.read_csv(GITHUB + FEATURES, sep='\t', encoding='utf-8')
features.info()
```

```python
PHONEMES ='/phoible/phoible/master/phoible-phonemes.tsv'

phonemes = pd.read_csv(GITHUB + PHONEMES, sep='\t', encoding='utf-8')
phonemes.info()
```

```python
nlangs = pd.DataFrame({'nlangs': phonemes.groupby('Phoneme').size()})
df = pd.merge(features, nlangs, 'left', left_on='segment', right_index=True)
df.set_index('segment', inplace=True)
df.sort('nlangs', ascending=False, inplace=True)
df
```

```python
feats = '''
labial
delayed_release
lateral
dorsal
distributed
epilaryngeal_source
constricted_glottis
raised_larynx_ejective
trill
consonantal
round
fortis
spread_glottis
stress
tap
low
periodic_glottal_source
nasal
labiodental
back
strident
retracted_tongue_root
click
lowered_larynx_implosive
approximant
tense
high
short
sonorant
continuant
long
syllabic
advanced_tongue_root
anterior
front
coronal
'''.strip().split()

rank = {f: i for i, f in  enumerate(feats)}
cols = sorted(df.columns, key=lambda c: rank.get(c, float('inf')))
df = df.reindex_axis(cols, axis=1)
df
```

```python
ct = df.iloc[:,:-1]
ct = ct[ct.isin(['+', '-', '0']).all(axis=1)]

plus = ct.apply(lambda v: v == '+').rename(columns=lambda c: '+' + c)
minus = ct.apply(lambda v: v == '-').rename(columns=lambda c: '-' + c)
cols = [c for p, m in zip(plus.columns, minus.columns) for c in (p, m)]

ct = pd.concat([plus, minus], axis=1).reindex_axis(cols, axis=1)
ct
```

```python
import concepts

def get_context(dataframe):
    objects = dataframe.index.values
    attributes = dataframe.columns
    bools = list(dataframe.itertuples(index=False))
    return concepts.Context(objects, attributes, bools)
    
c = get_context(ct.iloc[:10,:35])
c
```
